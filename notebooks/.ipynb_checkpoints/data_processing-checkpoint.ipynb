```python
from collections import Counter

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import chi2_contingency, chisquare, pearsonr
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from wordcloud import WordCloud
```

```python
post = Counter(df["V202073"])
print(post[1]+post[2])
```

```python
column_index = df.columns.get_loc('V202073')
for row in df.iterrows():
    if row[1]["V201029"] in (1,2): # and row[1]["V202073"] in (1,2):
        df.iloc[row[0], column_index] = row[1]["V201029"]

post = Counter(df["V202073"])
print(post[1]+post[2], post.keys())
```

```python
pres_data = []
pres_idx = []
for idx, val in enumerate(df["V202073"]):
    if val < 0 or val > 2:
        continue
    pres_data.append(val)
    pres_idx.append(idx)
```

```python
pres_labels = [1, 2], ["Joe Biden", "Donald Trump"]
plt.xticks(*pres_labels)
plt.hist(pres_data, bins=20)
```

```python
xdata = df["V201651"][pres_idx]

table = pd.crosstab(xdata, pres_data)
table = table[table.index > 0].T
# total_unique = len(np.unique(xdata))
xdata_bins = len([x for x in np.unique(xdata) if x > 0])

ax = sns.heatmap(table, annot=True, fmt=".0f", cmap="viridis")
plt.yticks(*pres_labels)
plt.xticks(
    range(1, xdata_bins + 1),
    reversed(["Very good", "Good", "Fair", "Poor", "Very poor"]),
)
# plt.title("When did R last work?")
plt.savefig("./outputs/when_did_last_work.png")
plt.show()

chi2, p, dof, expected = chi2_contingency(table)

print("Chi-Squared Test Results:")
print(f"Chi2 Statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of Freedom: {dof}")
print("\nExpected Frequencies (assuming no association):")
print(np.round(expected, 2))
```
